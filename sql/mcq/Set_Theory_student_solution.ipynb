```python
%%sql

SELECT COUNT(*) FROM Exammarks WHERE ( (Maths IS NULL) OR (Science IS NULL)
                               OR (Accounting IS NULL) OR (Biology IS NULL) OR (CompSci IS NULL));

```

```python
%%sql

SELECT COUNT(*) FROM Supplementarymarks WHERE ( (Maths IS NULL) OR (Science IS NULL)
                               OR (Accounting IS NULL) OR (Biology IS NULL) OR (CompSci IS NULL))
UNION 

SELECT COUNT(*) FROM Exammarks WHERE ( (Maths IS NULL) OR (Science IS NULL)
                               OR (Accounting IS NULL) OR (Biology IS NULL) OR (CompSci IS NULL));
```
