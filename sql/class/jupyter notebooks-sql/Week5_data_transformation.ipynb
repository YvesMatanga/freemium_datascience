<h4>String Manipulation functions of Interest</h4>
<ul>
    <li>LENGTH(str): Determine Length of a string</li>
    <li>REPLACE(string, pattern, replacement_string)</li>
    <li>TRIM(str)</li>
    <li>LTRIM(str)</li>
    <li>RTRIM(str)</li>
    <li>SUBSTR(str,star_index, num_chars)</li>
    <li>INSTR(str,substring): Return index of first occurence of substr</li>
    <li>UPPER(str)</li>
    <li>LOWER(str)</li>   
    <li>|| : Concatenate strings</li>
</ul>

<b>1. LENGTH</b>

```python
%%sql

SELECT
    Name,
    LENGTH(Name) AS NameLength,
    IDNumber,
    LENGTH(IDNumber) AS IDLength
FROM
    Students
LIMIT 5;
```

<b>2. REPLACE </b>

```python
%%sql

SELECT
    IDNumber,
    LENGTH(IDNumber) AS LengthOfID,
    REPLACE(IDNumber,'-','') AS IDNumberPreprocessed1, 
    REPLACE(REPLACE(IDNumber,'-',''),'#','') AS CorrectIDFormat, -- correct ID format such that we only have 13 characters in the string,
    LENGTH(REPLACE(REPLACE(IDNumber,'-',''),'#','')) AS LengthOfCorrectID
FROM
    Students
WHERE
    LengthOfID<>13
```

<b>3. TRIM </b>

--REGEX in SQL

```python
%%sql
SELECT Name, LENGTH(Name) AS NameLength, TRIM(Name) AS LeftTrimmedName,  (LENGTH(Name)-LENGTH(LTRIM(Name))) AS LeftSpaces FROM Students
LIMIT 10
```

<b>4. SUBSTR</b>

SUBSTR(str, start_pos, num_char)

9105106430182 -  YYMMDDxxxx(ddd)

```python
%%sql
SELECT
    Name,
    IDNumber,
    SUBSTR(IDNUmber,1,2) AS Year,
    SUBSTR(IDNUmber,3,2) AS Month,
    SUBSTR(IDNUmber,5,2) AS Day
FROM
    Students
LIMIT 5;
```

<b>5. INSTR</b>

-- return the index of first occurence of a string

```python
%%sql
SELECT
    Name,
    IDNumber,
    INSTR(IDNumber,'-') AS FirstOccurrence
FROM
    Students
WHERE LENGTH(IDNumber)>13
```

<b>6. LOWER, UPPER</b>

```python
%%sql
SELECT
    Name,
    UPPER(Name) AS Uppercase,
    LOWER(Name) AS Lowercase
FROM
    Students
LIMIT 5;  
```

<b>7. CONCATENATE: ||</b>

```python
%%sql

SELECT
    Name, Surname, Name ||' '|| Surname  AS FullName
FROM
    Students
LIMIT 5;
```

<b>2. CASE </b>

<img src="media/case.png"/>

```python
%%sql
SELECT
    Name,
    Title,
    CASE
        WHEN UPPER(Title) IN ('MS','MRS','MISS') THEN 'Female'
        WHEN UPPER(Title) IN ('MR') THEN 'Male'
        WHEN UPPER(Title) IS NULL THEN 'Value not specified'
    ELSE
        'Cannot Determine from Title'
    END AS Gender
FROM
    Employees
ORDER BY Name
LIMIT 5;
```

<b>3. IIF</b><br/>
IIF(condition_x,result_1,result_2)   -- if else

```python
%%sql 
SELECT * FROM Employees LIMIT 10;
```

```python
%%sql

SELECT
    Name,
    Title,
    IIF(UPPER(Title) IN ('MS','MRS','MISS'),'Female', IIF(UPPER(Title) IN ('MR'),'Male','Cannot Determine from Title'))AS Gender
FROM
    Employees
ORDER BY Name
LIMIT 5;
```

<b>4. COALESCE</b><br/>
return the first non-null argument value


```python
%%sql
SELECT
    
    Name,
    Surname,
    Level,
    Title,
    COALESCE(Title,'No title available') as Title_Coalesce
FROM Employees
LIMIT 5;
```

<b>5. NULLIF </b>

```python
%%sql
SELECT
    Name,
    Surname,
    Level,
    NULLIF(Level,'Intern') as Role
FROM Employees
WHERE Level = 'Intern'
```
