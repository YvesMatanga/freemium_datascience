```python
%%sql
SELECT BillingCountry AS 'Country', ROUND(SUM(Total),2) AS 'Revenue' FROM invoices GROUP by BillingCountry  ORDER BY Revenue DESC
```

<h1>Numeric Functions</h1>

```python
%%sql 
SELECT invoiceId, InvoiceDate, Total, ROUND(Total*20) AS 'Total in Rand' FROM invoices
```

<h1>String Functions</h1>

```python
%%sql
SELECT Title, LENGTH(Title) AS 'title_length' FROM albums
```

```python
%%sql
SELECT Title, UPPER(Title) as 'Title_Upper' FROM albums
```

```python
%%sql
SELECT Title, SUBSTRING(Title,1,3) FROM albums
```

```python
%%sql
SELECT Title||'hello' FROM Albums
```

<h1>Date Functions</h1>

```python
%%sql
SELECT DATE('now') AS 'today'
```

```python
%%sql
SELECT DATE('2050-08-21', '+10 days');
```

<h1>Conditional flow Functions</h1>

```python
%%sql

SELECT Total, IIF(Total > 2, 'expensive','cheap') AS Category FROM invoices
```

```python
%%sql
SELECT *FROM media_types
```

```python
%%sql
SELECT Name, 
CASE Name 
  WHEN 'MPEG audio file' THEN 'MPEG'
  WHEN 'AAC audio file' THEN 'AAC'
  ELSE 'None'
END AS 'Type'
FROM media_types
```
