<b>UPDATE</b>: Modify a record or a set or records using a given criterion

```python
%%sql

UPDATE customer_list 
SET email = 'yves_matanga@gmail.com', tel='0873655461' WHERE name ='yves';
```

DELETE: remove a record or a set of records using a given criterion

<b>DELETE FROM table_name WHERE criterion</b>

```python
%%sql

DELETE FROM customer_list WHERE name='yves';
```

<h2>Sorting and Aggregate Functions</h2>
<ul>
    <li>SUM, MIN, MAX</li>
    <li>COUNT</li>
    <li>ORDER BY</li>
    <li>LIMIT</li>
    <li>GROUP BY</li>
    <li>HAVING  (WHERE statements used in a group by query)</li>
</ul>


```python
%%sql

SELECT COUNT(*) AS 'num of products', MIN(PRICE) AS 'minimum price',  MAX(price) AS 'maximum price'
, SUM(price)/COUNT(*) AS 'average price' FROM prd_list;
```

```python
%%sql

SELECT * FROM prd_list WHERE (price < 4000) ORDER BY price ASC;
```

```python
%%sql 

SELECT * FROM invoice_details;
```

```python
%%sql

SELECT inv_id, COUNT(*), SUM(qty) AS 'quantity_per_invoice', SUM(qty)/COUNT(*) AS 'average quantity per invoice' FROM invoice_details GROUP BY inv_id
HAVING (COUNT(*) > 3); 
```

```python
%%sql

SELECT * FROM invoice_details INNER JOIN invoice ON invoice_details.inv_id == invoice.invoice_id
```

```python
%%sql

SELECT * FROM prd_list WHERE price < (SELECT SUM(price)/COUNT(*) FROM prd_list);
```
