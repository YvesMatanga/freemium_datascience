<b>Question 6</b><br/><br/>
Find the customers (with no duplicates) who have made a purchase (appear in the invoice table).

Find the total revenue generated from all invoices.

<details style="color:blue">
    <summary>Click to reveal the <b>answer<b/></summary>

  **Answer:**
        SELECT SUM(invoice_details.qty*prd_list.price) AS 'total revenue' FROM invoice_details JOIN prd_list ON invoice_details.prd_id = prd_list.id

<b>Question 9</b><br/><br/>
List the customers who made a purchase between June and July.

```python
%%sql 

SELECT COUNT(*) FROM customer_list WHERE customer_list.id IN (SELECT invoice.cust_id FROM invoice WHERE date >= '2018-06-01'
                                                      AND date <= '2018-07-31')
```
