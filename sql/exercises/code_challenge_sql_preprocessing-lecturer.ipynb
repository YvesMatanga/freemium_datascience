<b>Tax Brackets</b>

Annual Tax: 

0% on income up to the tax threshold<br/>
18% on income between the threshold and R205,900<br/>
26% on income between R205,901 and R321,600<br/>
31% on income between R321,601 and R445,100<br/>
36% on income between R445,101 and R584,200<br/>
39% on income between R584,201 and R744,800<br/>
41% on income between R744,801 and R1,577,300<br/>
45% on income above R1,577,301<br/>

Annual Tax Rebate: R14958<br/>
Annual Tax Threshold: R87300

<b>Example</b>
<p style="color:blue">Assumptions:</p>

Monthly salary: R25,000
Age: Under 65 years old
No additional deductions or credits for simplicity
Determine the Taxable Income:

<p style="color:blue">Apply Tax Threshold:</p>

Monthly threshold: R87,300 / 12 = R7,275
If the monthly income is below this threshold, there is no income tax payable. In this example, R25,000 > R7,275.

<p style="color:blue">Apply Tax Brackets:</p>

The portion of income up to R205,900 is taxed at 18%.
The portion of income between R205,901 and R321,600 is taxed at 26%.
In this example, the entire income falls within the first bracket (18%).

<p style="color:blue">Tax Calculations:</p>

Tax on the portion of income in the first bracket: R25,000 * 18% = R4,500
Determine Rebates:

<p style="color:blue">Apply Tax Rebate:</p>

Monthly rebate: R14,958 / 12 = R1,246.50

Subtract the rebate from the tax calculated:

Net Tax: R4,500 - R1,246.50 = <b style="color:red">R3,253.50</b>
