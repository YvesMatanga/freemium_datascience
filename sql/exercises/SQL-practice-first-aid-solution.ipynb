<h1>Example: Tax Calculation Breakdown</h1>

<p class="blue">Assumptions:</p>
<ul>
    <li>Monthly salary: R25,000</li>
    <li>Age: Under 65 years old</li>
    <li>No additional deductions or credits for simplicity</li>
</ul>

<h3 class="blue">1. Determine the Taxable Income:</h3>
<p>Annual salary: R25,000 x 12 = <span class="bold">R300,000</span></p>

<h3 class="blue">2. Apply Tax Threshold:</h3>
<p>Annual tax threshold: <span class="bold">R87,300</span></p>
<p>Since R300,000 is above this threshold, income tax applies.</p>

<h3 class="blue">3. Apply Tax Brackets:</h3>
<ul>
    <li>First R205,900 taxed at 18%</li>
    <li>Next portion between R205,901 and R321,600 taxed at 26%</li>
</ul>

<p>For R300,000:</p>
<ul>
    <li>Tax on first R205,900: R205,900 x 18% = <span class="bold">R37,062</span></li>
    <li>Tax on remaining R94,100 (R300,000 - R205,900) at 26%: R94,100 x 26% = <span class="bold">R24,466</span></li>
</ul>

<h3 class="blue">4. Calculate Total Tax Before Rebates:</h3>
<p>Total Tax = R37,062 + R24,466 = <span class="bold">R61,528</span></p>

<h3 class="blue">5. Apply Annual Tax Rebate:</h3>
<p>Annual rebate = R14,958</p>
<p>Tax after rebate = R61,528 - R14,958 = <span class="bold">R46,570</span></p>

<h3 class="blue">6. Monthly Tax Calculation:</h3>
<p>Monthly tax = R46,570 รท 12 = <span class="bold">R3,880.83</span></p>

<h3 class="blue">7. Net Salary Calculation:</h3>
<p>Net salary = R25,000 - R3,880.83 = <span class="bold blue">R21,119.17</span></p>

<h3 class="blue">Summary:</h3>
<ul>
    <li><span class="blue">Annual Taxable Income</span>: R300,000</li>
    <li><span class="blue">Annual Tax before Rebate</span>: R61,528</li>
    <li><span class="blue">Annual Tax after Rebate</span>: R46,570</li>
    <li><span class="blue">Monthly Tax</span>: R3,880.83</li>
    <li><span class="blue">Net Monthly Salary</span>: <span class="bold blue">R21,119.17</span></li>
</ul>
