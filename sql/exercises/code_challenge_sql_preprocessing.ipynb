<b>Question 1</b><br/><br/>
Given the Employees Table in the SoftDevEmployees database, generate a result set with the following columns and requirements:
<ul>
    <li>Title: Must be uniform (example: Mr, MR or mR must be normalised). If the title is not given change it to unknown</li>
    <li>Name: Must have not leading or trailing space - First letter must be in upper case</li>
    <li>Surname: Must have no leading or trailing spaces - First letter must be in upper case</li>
    <li>Gender: The gender must be inferred from the Title</li>
</ul>

Save the newly retrieved data into an excel sheet. Hint: Use Pandas export.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../SoftDevEmployees.db')
query = """
SELECT IIF(Title IS NULL,'Unknown',UPPER(SUBSTR(Title,1,1))||LOWER(SUBSTR(Title,2,LENGTH(Title)-1))) AS Title,
    TRIM(Name) AS Name,
    TRIM(Surname) AS Surname,
    CASE 
        WHEN UPPER(Title) IN ('MISS', 'MS', 'MRS') THEN 'Female'
        WHEN UPPER(Title) ='MR' Then 'Male'
        ELSE 'Unknown'
    END As Gender
FROM Employees
"""
df = pd.read_sql_query(query,conn)
conn.close()

df.head()
df.to_excel("result.xlsx",index=False)
```

<b>Question 2</b><br/><br/>
Given the Employees Table in the SoftDevEmployees database, generate a result set with the following columns and requirements:
<ul>
    <li>Title: Must be uniform (Mr, MR or mR must be normalised)</li>
    <li>Name: Must have not leading or trailing space - First letter must be in upper case</li>
    <li>Surname: Must have no leading or trailing spaces - First letter must be in upper case</li>
    <li>Gender: The gender must be inferred from the Title</li>
    <li><b>Salary</b>: Assuming that the salary column is the gross salary</li>
    <li><b>Tax Deductions</b>: See process in cell below</li>
</ul>
Save the newly retrieved data into an excel sheet. Hint: Use Pandas export.

```python
%%sql

SELECT 
 Salary, IIF( Salary < 7275, 0,            
             CASE
                 WHEN Salary*12 < 205900 THEN CAST(Salary*0.18-14958/12  AS INT)
                 WHEN Salary*12 < 321600 THEN CAST(Salary*0.26-14958/12  AS INT)
                 WHEN Salary*12 < 445100 THEN CAST(Salary*0.31-14958/12  AS INT)
                 WHEN Salary*12 < 584200 THEN CAST(Salary*0.36-14958/12 AS INT)
                 WHEN Salary*12 < 744800 THEN CAST(Salary*0.39-14958/12  AS INT)
                 WHEN Salary*12 < 1577300 THEN CAST(Salary*0.41-14958/12 AS INT)
                 ELSE Salary*0.45-14958/12              
             END
            ) AS TaxDeductions, 
     IIF( CAST(Salary AS REAL) < 7275, Salary,            
             CASE
                 WHEN Salary*12 < 205900 THEN CAST( Salary - (Salary*0.18-14958/12 ) AS INT)
                 WHEN Salary*12 < 321600 THEN CAST( Salary - (Salary*0.26-14958/12 ) AS INT)
                 WHEN Salary*12 < 445100 THEN CAST(Salary - (Salary*0.31-14958/12 ) AS INT)
                 WHEN Salary*12 < 584200 THEN CAST(Salary - (Salary*0.36-14958/12 ) AS INT)
                 WHEN Salary*12 < 744800 THEN CAST(Salary - (Salary*0.39-14958/12 ) AS INT)
                 WHEN Salary*12 < 1577300 THEN CAST(Salary - (Salary*0.41-14958/12 ) AS INT)
                 ELSE CAST(Salary*0.45-14958/12 AS INT)
             END
            ) AS NetSalary
    FROM Employees ORDER BY TaxDeductions ASC LIMIT 20
```
