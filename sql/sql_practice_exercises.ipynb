        SELECT DISTINCT c.* FROM customer_list AS c INNER JOIN invoice ON invoice.cust_id = c.id
</details>

<b>Question 7</b><br/><br/>
Retrieve the customer fullnames who have not made any purchases.

<details style="color:blue">
    <summary>Click to reveal the <b>answer<b/></summary>

  **Answer:**
        SELECT c.* FROM customer_list AS c WHERE c.id NOT IN (SELECT invoice_id FROM invoice)
</details>

Find the total revenue generated from all invoices.

<details style="color:blue">
    <summary>Click to reveal the <b>answer<b/></summary>

  **Answer:**
        SELECT SUM(invoice_details.qty*prd_list.price) AS 'total revenue' FROM invoice_details JOIN prd_list ON invoice_details.prd_id = prd_list.id
</details>
