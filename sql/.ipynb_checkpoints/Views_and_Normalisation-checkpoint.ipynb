<h2>Creating a view</h2><br/>
The syntax to create a view is 

<b>CREATE VIEW</b> View_Name <b>AS</b>

 --- SELECT xxxxx

<b>Problem 1</b><br/>
Create a view that will display all the invoices in the database, the name and surname of their owners, and the date of the invoice.

```python
%%sql
DROP VIEW Customer_View;

CREATE VIEW Customer_View AS
SELECT * FROM customer_list;
```

```python
%%sql
SELECT * FROM Customer_View
```

```python
%%sql
DROP VIEW IF EXISTS Invoice_View;

CREATE VIEW Invoice_View AS
SELECT invoice_id AS 'InvoiceID', name, surname, date
FROM invoice JOIN customer_list ON invoice.cust_id = customer_list.id;
```

```python
%%sql 
SELECT * FROM Invoice_View;
```

<h2>2. Normalisation</h2><br/>
Normalisation is a database technique for evaluting and reorganising table structures to <br/>
<ul>
    <li>Minimise redundancies </li>
    <li>Improve data integrity</li>
</ul>

Normalisation can be segmented in stages based on how well redundancy is minimised:<br/>
<ul>
    <li>First Normal Form (<b>1NF</b>) </li>
    <li>Second Normal Form (<b>2NF</b>)</li>
    <li>Third Normal Form (<b>3NF</b>)</li>
    <li>Boyce-Codd Normal Form (<b>BCNF</b>)</li>
    <li>Fourth Normal Form (<b>4NF</b>)</li>
</ul>



<h3>1NF: Remove repeating groups</h3>

<h3>2NF: Remove Partial redundancies</h3>

<h3>3NF: Remove Nonprimary-key dependent columns</h3>

<h3>Normalisation process</h3>
<ul>
    <li>List all the attributes/columns, <b>grouping related data</b> together: 0NF</li>
    <li>Identify a Primary</li>
    <li>Perform 1NF: Remove repeating groups/Identify many-to-many relationships</li>
    <li>Perform 2NF: Remove part-key dependencies/Solve many-to-many relationships</li>
    <li>Perform 3NF: Remove inter-data dependencies/Resolve one-to-many relationships</li>    
</ul>
