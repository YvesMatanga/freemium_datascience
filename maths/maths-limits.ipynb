<h1>Limits at Infinity</h1>

$$
\lim{x\to\infty}\sqrt{9x^2+x}-3x
$$

$$
 =\lim{x\to\infty}\frac{(\sqrt{9x^2+x}-3x)(\sqrt{9x^2+x}+3x)}{\sqrt{9x^2+x}+3x}
$$

$$
 =\lim{x\to\infty}\frac{(9x^2+x-9x^2)}{\sqrt{9x^2+x}+3x}
$$

$$
 =\lim{x\to\infty}\frac{x}{\sqrt{9x^2+x}+3x}
$$

$$
 =\lim{x\to\infty}\frac{\frac{x}{x}}{\frac{\sqrt{9x^2+x}+3x}{x}}, x = \sqrt{x^2} , x > 0
$$

$$
\lim{x\to\infty}\frac{\frac{x}{x}}{\frac{\sqrt{9x^2+x}}{x}+\frac{3x}{x}}$$

$$
\lim{x\to\infty}\frac{\frac{x}{x}}{\frac{\sqrt{9x^2+x}}{\sqrt{x^2}}+\frac{3x}{x}}$$

<b>Problem 2</b>

$$
\lim{x\to\infty} \sqrt{x^2+ax} - \sqrt{x^2+bx}
$$

$$
= \lim{x\to\infty} \frac{(\sqrt{x^2+ax} - \sqrt{x^2+bx})(\sqrt{x^2+ax} + \sqrt{x^2+bx})}{(\sqrt{x^2+ax} + \sqrt{x^2+bx})}
$$

$$
= \lim{x\to\infty} \frac{(x^2+ax)-(x^2+bx)}{(\sqrt{x^2+ax} + \sqrt{x^2+bx})}
$$

$$
= \lim{x\to\infty} \frac{(x^2+ax-x^2-bx)}{(\sqrt{x^2+ax} + \sqrt{x^2+bx})}
$$

$$
= \lim{x\to\infty} \frac{(ax-bx)}{(\sqrt{x^2+ax} + \sqrt{x^2+bx})}
$$

$$
= \lim{x\to\infty} \frac{\frac{(ax-bx)}{x}}{\frac{(\sqrt{x^2+ax} + \sqrt{x^2+bx})}{x}}
$$

$$
= \lim{x\to\infty} \frac{a-b}{\frac{\sqrt{x^2+ax}}{x}+\frac{\sqrt{x^2+bx}}{x}}
$$

$$
= \lim{x\to\infty} \frac{a-b}{\frac{\sqrt{x^2+ax}}{\sqrt{x^2}}+\frac{\sqrt{x^2+bx}}{\sqrt{x^2}}}
$$

$$
= \lim{x\to\infty}\frac{a-b}{\sqrt{\frac{x^2+ax}{x^2}}+\sqrt{\frac{x^2+bx}{x^2}}}
$$

$$
= \lim{x\to\infty}\frac{a-b}{\sqrt{1+\frac{a}{x}}+\sqrt{1+\frac{b}{x}}}
$$

$$
=\lim{x\to\infty}\frac{a-b}{2} = \frac{a-b}{2}
$$

<b>Problem 3</b>

Find the asymptotes:

$$
f(x) = \frac{5+4x}{x+3}
$$

Vertical asymptote: Find x = a such that
$$
\lim{x\to a}f(x)=\infty
$$

Horizontal asymptote: Find y = L such that
$$
\lim{x\to\infty}f(x)=L
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x_n = np.arange(-11,-3.3,step=0.01)
f_n = (5+4*x_n)/(x_n+3)

z_n = np.arange(-2.9, 5, step=0.01)
g_n = (5+4*z_n)/(z_n+3)

plt.figure()
plt.plot(x_n,f_n)
plt.plot(z_n,g_n)
plt.show()
```
