<h3>Lesson 1: List</h3>
<p>A list is a collection of data types (objects) organised in an orderly fashion</p>
<p>Data can be inserted and retrieved by means of <b>Indexes</b></p>
<p>Lists are <b>mutable</b> objets, meaning its content can be <b>changed</b></p>
<p>A List can be shorten or expanded </p>

Functions and methods to remember: <b>len()</b>, <b>.append()</b>, <b>.remove()</b>

```python
my_list = [1,2,3,4,5] #creating a list of 5 elts
print("Element 1:", my_list[0])
print("Element 2:", my_list[1])

my_list2 = [i for i in range(5)]
print(my_list2)
```

<h4>Lesson 1b: List comprehension</h4>
<p>List comprehension refers to creating a list on the basis of another list. A short form syntax exists to achieve such</p>

syntax: <b>newlist = [expression for item in iterable if condition == True]</b>

```python
my_list_a = [3, 4, 6, 8]
my_list_b = [item for item in my_list_a if item%2==0]

print(my_list_b)
```

<h3>Lesson 3: Dictionnary</h3>
<p>A dictionary is a collection of data types stores by means of unique identifiers called <b>"keys"</b></p>
<p>A Dictionary is similar to a list, however, data are accessed by <b>keys</b> rather than <b>indexes</b></p>
<p><b>car_description[0] = 'mazda'</b>  (using a list)</p>
<p><b>car['brand'] = 'mazda'</b>  (using a dictionary)</p>
<p>It is appropriate to use a dictionary when storing an object with different properties</p>
<p>e.g. <b>soccer_player = {"name":"lionel", "surname":"messi", "age":34}<b/></p>
<p><b>player_name = soccer_player['name']</b></p>
