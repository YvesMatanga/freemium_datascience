<b>3.3 Recursive functions </b>

A <b>recursive function</b> is a function that <b>calls itself in order to solve a problem</b>. The key idea behind recursion is to break down a complex problem into smaller, more manageable subproblems of the same type, until reaching a base case that stops the recursion.

<b>Key Components of a Recursive Function:</b><br/><br/>
<b>1. Base Case</b>: The condition where the recursion stops. Without a base case, the recursion would continue indefinitely, leading to a stack overflow error.<br/>
<b>2. Recursive Case</b>: The part of the function where the function calls itself with a modified argument to break down the problem into smaller parts.

factorial (5) = 5! = 5 x 4 x 3 x 2 x 1 =  5 x 4!= 5 x 4 x 3! <br/>
1! = 1<br/>
0! = 1


factorial(n) = n*factorial(n-1)

```python
def factorial(n):
    # Base case: when n is 1, stop the recursion
    if n == 1:
        return 1
    else:
        # Recursive case: call factorial on n-1
        return n * factorial(n - 1)
```

```python
factorial(5)
```

<h2>5. Classes and Objects</h2>

Class = blueprint of a spieces sharing a given skeleton (abstract)<br/>
Object = actual instance of the class (tangible and concrete)

<br/>
<img width="500px" src="media/class_vs_object.png"/>

```python
class Dog: #PEP8
    # Constructor to initialize attributes of a dog
    def __init__(self, name, breed, age):
        self.name = name   # Assign the name
        self.breed = breed # Assign the breed
        self.age = age     # Assign the age

    # Method to make the dog bark
    def bark(self):
        return f"{self.name} says: Woof!"

    # Method to display dog info
    def dog_info(self):
        return f"Name: {self.name}, Breed: {self.breed}, Age: {self.age} years old"

    # Method to calculate dog birth date
    def birth_year(self):
        return 2024-self.age
```

```python
# Creating instances (objects) of the Dog class
dog1 = Dog(name="Buddy", breed="Golden Retriever", age=5)
dog2 = Dog(name="Max", breed="Bulldog", age=3)
```

```python
# Calling methods on the objects
print(dog1.bark())                 # Output: Buddy says: Woof!
print(dog1.dog_info())             # Output: Name: Buddy, Breed: Golden Retriever, Age: 5 years old
print(f"{dog1.name}'s birth year is {dog1.birth_year()}.")
```

```python
# Calling methods on the objects
print(dog2.bark())                 # Output: Buddy says: Woof!
print(dog2.dog_info())             # Output: Name: Buddy, Breed: Golden Retriever, Age: 5 years old
print(f"{dog2.name}'s birth year is {dog2.birth_year()}.")
```
