<h2>3. Python arithmetics</h2>

Basic Arithmetics:
<ul>
<li>Addition (+)</li>
<li>Subtraction (-)</li>
<li>Multiplication (*)</li>
<li>Division (/)</li>
<li>Integer Division (//)</li>
    <li>Exponent (**)</li>
</ul>

Additional functions: round, math.floor, math.ceil

```python
num_1 = 10
num_2 = 3

sum = num_1 + num_2
diff = num_1 - num_2
prod  = num_1 * num_2
quot = num_1/num_2 
iquot = num_1//num_2
exp = num_1**num_2
rem = num_1%num_2

print("sum",sum)
print("diff",diff)
print("product",prod)
print("quotient",quot)
print("integer division",iquot)
print("exponent",exp)
print("remainder",rem)
```

```python
import math

discount = 1.26
discount_floor = math.floor(discount)
discount_ceil = math.ceil(discount)
discount_res = round(discount,1)
print(discount_res,"-",discount_floor,"-", discount_ceil)
```

<h2>3.2 String Manipulation</h2>

Basic string manipulation functions/commands: 
<ul>
    <li>len</li>
    <li>lower/upper</li>
    <li>replace</li>
    <li>split</li>
    <li>strip</li>
</ul>

```python
poem = '       I see a beautiful city rising from this abyss         '
r_poem = poem.replace('see','view')
words = poem.split(' ')
poem_denoising = poem.strip()

num_characters = len(poem)
print(f'Poem size: {num_characters}')
print(f'Case folding: {poem.upper()}')
print(f'Replacement {r_poem}')
print(words)
print("[",poem_denoising,"]",end="")
```

<h4>Condition = Conditional code<h4>
    <p>If statement</p>
    <p>If Else statement</p>
    <p>If Elif ELif ...Else statement</p>

```python
# age < 5: baby
# 5 - 15: teenager
# 15 - 18: adulting teenager
#18 - 35: adult
#35 - 65: mature adult
#65 - 75 : very mature adult


age = int(input('What is age?'))

if age < 5:
    print('You are a baby')
elif age < 15:
    print("You are a teenager")
elif age < 18:
    print("You are an adulting teeanager")
elif age < 35:
    print("You are an adult")
elif age < 65: 
    print("You are a mature adult")
elif age < 75:
    print("You are very mature adult")
else:
    print("You are an elderly")
```

<h2>5. Python data structures</h2>

<p>Data Structures are used to <b>store user data using a specific data organisation scheme</b> that <b>yields</b> some processing or data retrieval <b>benefits</b> </p>

<ol>
        <li>Lists/Arrays</li>
        <li>Tuples</li>
        <li>Dictionaries</li> 
        <li>Sets</li>
        <li>Special methods and functions: .append(), .insert(),.extend(), len(), del, .pop(), .clear(), foreach, inline for, zip</li>
    </ol>  

<b>A list is a collection of data types (objects) organised in an orderly fashion</b><br/>
<p>Data can be inserted and retrieved by means of <b>Indexes</b></p>
<p>Lists are <b>mutable</b> objets, meaning its content can be <b>changed</b></p>
<p>A List can be shorten or expanded </p>

Functions and methods to remember: <b>len()</b>, <b>.append()</b>, <b>.remove()</b>

<p><b>A tuple is collection of data types organised in orderly fashion</b></p>
<p>Unlike Lists, <b>Tuples are immutable objects</b>, that is its content cannot be modified </p>
<p>Functions and methods to remember: <b>len()</b></p>

<p><b>A set is a collection of unique objets, where duplicates data types are not allowed</b></p>
<p><b>fruits = {"banana","orange","mango", "mango"}</b> 
    <br/>will be stored as <b><br/>fruits = {"banana","orange","mango"}</b></p>

<p><b>A dictionary is a collection of data types stores by means of unique identifiers</b> called <b>"keys"</b></p>
<p>A Dictionary is similar to a list, however, data are accessed by <b>keys</b> rather than <b>indexes</b></p>
<p><b>car_description[0] = 'mazda'</b>  (using a list)</p>
<p><b>car['brand'] = 'mazda'</b>  (using a dictionary)</p>
<p>It is appropriate to use a dictionary when storing an object with different properties</p>
<p>e.g. <b>soccer_player = {"name":"lionel", "surname":"messi", "age":34}<b/></p>
<p><b>player_name = soccer_player['name']</b></p>
