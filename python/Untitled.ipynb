<h4>If Else Statement</h4>

```python
age = int(input('Enter your age:'))

if age < 18:
    print("You are a minor")
    print("You have a long life")
else:
    print("You are an adult")
    print('Start taking your life seriously!')
    
```

<h4>If Elif Elif Elif....... Else Statement</h4>
< 18: A minor<br/>
>= 18 and 35 <= Young adult<br/>
35 and 60 < You are mature adult<br/>
60 and 75 < You are very mature adult<br/>
> 75: Elderly

```python
age = int(input('Enter your age:'))

if age < 18: 
    print("You are minor")
elif (age >= 18) and (age <= 35):
    print("You are a young adult")
elif (age > 35) and (age <= 60):
    print("You are a mature adult")
elif (age > 60) and (age <= 75):
    print("You are a very mature adult")
else:
    print("You are an elderly")
```
