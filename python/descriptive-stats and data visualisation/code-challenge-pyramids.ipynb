```python
def print_pyramid(height):
    for i in range(height):
        spaces = " " * (height - i - 1)
        stars = "*" * (2 * i + 1)
        print(f"{spaces}{stars}")
```

```python
print_pyramid(20)
```

```python
def dynamic_triangle_me_v(num):
    odd_star_count = 1
    for l in range(num): #print layers
        #print spaces
        for i in range(num-l-1):
            print(" ",end="")
        #print the stars
        for i in range(odd_star_count):
            print("*",end="")
        print("")
        odd_star_count +=2
        
```

```python
dynamic_triangle_me_v(12)
```
