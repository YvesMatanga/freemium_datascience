<h3>1. Matrices and Vectors</h3>

A <b>vector</b> is a set of values. It is a multi-dimensional representation of a scalar value, in basic algebra.

$$
 x = \begin{bmatrix}
 x_1\\
 x_2\\
 x_3\\
 \end{bmatrix}
$$

<h4>Data representation - view point</h4>

How can we represent a person's profile, with an age of 20, a height of 1.86m, a weight of 86.3Kg and three children?

$$
p = \begin{bmatrix}
 20\\
 1.86\\
 86.3\\
 3\\
 \end{bmatrix}
$$

A <b>matrix</b> is a set of values organised in grid format. Fundementally, in linear algebra, it is used to represent linear transformation. However, in general practice it can be used as an entity to store values.

<h4>Data representation - view point</h4>

How will we represent several observations of people with the characteristics described in the previous example?

$$
m = \begin{bmatrix}
 20 & 1.86 & 86.3 & 3\\
 25 & 1.45 & 75.6 & 4\\
    23 & 1.5 & 65 & 3\\
    21 & 1.73 & 85 & 0\\
    19 & 1.43 & 81.5 & 0\\
 \end{bmatrix}
$$

A matrix is defined by its number of rows and columns. We say the <b>matrix m is a matrix of 5 by 4 </b>($m_{5x4} $)

<h4>Algebra - view point</h4>

<p>Matrices have got a different connation in linear algebra. A matrix in linear algebra, it is a set of transformation coefficient to modify a vector.</p>

$$
A = \begin{bmatrix} a_{11} & a_{12}\\a_{21} & a_{22}\end{bmatrix}, 
   b = \begin{bmatrix}
    b_1\\
    b_2\\
    \end{bmatrix}
$$

$$
c = A.b = b_1\begin{bmatrix}a_{11}\\a_{21}  \end{bmatrix} + b_2\begin{bmatrix}a_{12}\\a_{22} 
\end{bmatrix} = \begin{bmatrix}a_{11}b_1 +a_{12}b_2\\a_{21}b_1 +a_{22}b_2\\ \end{bmatrix}
$$

For instance, a matrix
$$
 A = \begin{bmatrix}
 2 & 0\\ 0 & 3 
 \end{bmatrix}
$$
will multiply the first value of any two-valued vector by 2 and the second by 3.

$$
 x = Ab = \begin{bmatrix}
 2 & 0\\ 0 & 3 
 \end{bmatrix}b = \begin{bmatrix}
 2 & 0\\ 0 & 3 
 \end{bmatrix}\begin{bmatrix}b_1\\b_2\end{bmatrix} = \begin{bmatrix}
 2b_1& 0\\ 0 & 3b_2 
 \end{bmatrix} = \begin{bmatrix}2b_1\\3b_2\end{bmatrix}
$$

<h4>The transpose of a matrix</h4>

The transpose of a matrix $A = [a_{ij}] $ of size $m x n$ is a matrix $A^{T} = [a_{ji}] $ if size $ n x m$

$$
m = \begin{bmatrix}
 20 & 1.86 & 86.3 & 3\\
 25 & 1.45 & 75.6 & 4\\
    23 & 1.5 & 65 & 3\\
    21 & 1.73 & 85 & 0\\
    19 & 1.43 & 81.5 & 0\\
 \end{bmatrix}
$$

$$
m^T = \begin{bmatrix}
 20 & 25 & 23 & 21 & 19\\
 1.86 & 1.45 & 1.5 & 1.73 & 1.43\\
    86.3 & 75.6 & 65 & 85 & 81.5\\
    3 & 4 & 3 & 0 & 0\\
 \end{bmatrix}
$$

<ul>

</ul>

```python
import warnings 
warnings. filterwarnings('ignore')

import pandas as pd

df = pd.read_csv('datasets/Healthcare-Diabetes.csv') #loading data into a dataframe

def age_grouping(age):
    if age < 20:
        group = 'teenager'
    elif age < 40:
        group = 'adult'
    else:
        group = 'old-person'
    return group

df['age-group'] = df['Age'].apply(age_grouping)
df.head()
```

<h4>Boxplot</h4> <br/>

Summarises the variable

```python
import matplotlib.pyplot as plt

plt.figure()
plt.boxplot(df['Age'])
plt.title('Box plot - Age variable')
plt.show()
```

<b>The Probability density function: pdf(x) = P(x)</b>

```python
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

density_dist = gaussian_kde(df['Age'])


x_min = df['Age'].min()
x_max = df['Age'].max()
r_x = x_max - x_min
x_n = np.arange(x_min,x_max,step=r_x/100)


d_x = density_dist(x_n) #estimtate the probability density function of the data

plt.figure()
plt.plot(d_x)
plt.xlabel('Age')
plt.ylabel('Density function')
plt.show()
```
