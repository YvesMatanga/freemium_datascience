<h3>Exercise 3</h3>

$$
BMI = \frac{mass (Kg)}{height (m^2)}
$$
<p>
    <table>
        <thead>

        </thead>

        <tr>
            <td> Between 25 and 29.99 </td><td> Overweight</td>
        </tr>
        <tr>
            <td> $\geq 18.5$ </td><td> Obese</td>

</p>
