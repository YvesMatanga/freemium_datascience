```python
def menu():
    print("*"*15)
    print("|Calculator|")
    print("*"*15)
    print("|1. Add")
    print("|2. Subtract")
    print("|3. Exit")
    print("*"*15)
```

```python
exit = False
while (not exit):
    menu()
    choice = int(input('Make your Choice:'))    
    if choice == 1:
        print("*"*15)
        print("|Addition ")
        print("*"*15)
        num_1 = float(input('Enter number 1:'))
        num_2 = float(input('Enter number 2:'))
        res = num_1 + num_2
        print(num_1,"+",num_2,"=",res)
        input("Press any key to continue...")
    elif choice == 2:
        print("*"*15)
        print("|Subtraction ")
        print("*"*15)
        num_1 = float(input('Enter number 1:'))
        num_2 = float(input('Enter number 2:'))
        res = num_1 - num_2
        print(num_1,"-",num_2,"=",res)
        input("Press any key to continue...")
    elif choice == 3:
        exit = True
    else:
        print("Incorrect choice")    
```
