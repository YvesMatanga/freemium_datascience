<h1>Basic Arithmetic</h1>

```python
number1 = 20
number2 = 3

sum = number1 + number2 #add
diff = number1 - number2#subtract
quotient = number1 / number2#divide
product = number1*number2  #multiply

rem = number1%number2 #finding the remainder of integer division
int_div = number1//number2 #integer division
power = number1**number2 #number1 to the power of number 2

print(f"{number1} + {number2} = {sum}")
print(f"{number1} - {number2} = {diff}")
print(f"{number1} * {number2} = {product}")
print(f"{number1} / {number2} = {quotient}")
print(f"{number1} mod {number2} = {rem}")
print(f"{number1} // {number2} = {int_div}")
print(f"{number1}^{number2} = {power}")
```

<h1>Routine Functions</h1>

```python
import math
#round 
number = 20.66667
number_rounded  = round(number, 2)

print(number_rounded)

#floor
number_truncated = math.floor(number)
#ceil
number_roundup = math.ceil(number)

print(number_truncated)
print(number_roundup)
```

<h1>String Manipulation</h1>

```python
#len: count the number of characters

name = 'andile'
num_char = len(name)
print(f"number of characters in {name}: {num_char}")

#concatenate: join strings

name1 = 'kim'
name2 = 'ukukhanya'

full_name = name1+' '+name2
print(full_name)
#lower case
name3 = 'WORLD CUP'
name_lowered = name3.lower()
print(name_lowered)
#upper case
name_upper_case = name1.upper()

print(name_upper_case)

#replace 
key = 'e'
change = 'a'
word = 'hello everyone'
replaced_word = word.replace(key, change)#returns a new word with replacements
print(replaced_word)

#strip: is a function to remove spaces on the left and right hand side of the word
word2 = ' I am poorly formatted   '
proper_word = word2.strip()   #rstrip remove space to the right , lstrip : remove to the left
print("#"+proper_word+"#")

#split(delimiter) : string delimiter
long_text = 'The world is changed'
sep = ' '
words = long_text.split(sep)
print(words)
```
