<h3>Lesson 1: List</h3>
<p>A list is a collection of data types (objects) organised in an orderly fashion</p>
<p>Data can be inserted and retrieved by means of <b>Indexes</b></p>
<p>Lists are <b>mutable</b> objets, meaning its content can be <b>changed</b></p>
<p>A List can be shorten or expanded </p>

Functions and methods to remember: <b>len()</b>, <b>.append()</b>, <b>.remove()</b>

```python
my_list = [1, 2, 'hello', 5.6]
print(my_list)
print(type(my_list))

first_elt = my_list[0] #indexing
last_elt  = my_list[3] #last index

print("first element: ",first_elt)
print("last element:", last_elt)
```

<h4>Accessing Elements in a list</h4>

```python
n = len(my_list) #the number of elements

for i in range(n):#0, 1, 2, ,n-1  e.g range(3) = 0,1,2
    elt = my_list[i]#fetch the element    
    print(f" elt - {i+1}: {elt}")

#f = "formatted print"
# f"...........  {variable_content}"
```

```python
i = 0
for elt in my_list: #foreach "reading": foreach element in my_list
    print(i+1,"element:", elt)
    i = i + 1
```

```python
print(my_list)
last_elt = my_list[-1]
before_last_elt = my_list[-2]

print(last_elt)
print(before_last_elt)
```

```python
sub_list = my_list[0:1]
sub_list2 = my_list[-3:-1]
print(sub_list)
print(sub_list2)
```

```python
sub_list3 = my_list[1:]
sub_list4 = my_list[:2]

print(sub_list3)
print(sub_list4)
```

```python
my_list = [1,2,3,4,5] #creating a list of 5 elts
print("Element 1:", my_list[0])
print("Element 2:", my_list[1])

my_list2 = [i for i in range(5)]
print(my_list2)
```

<h4>Lesson 1b: List comprehension</h4>
<p>List comprehension refers to creating a list on the basis of another list. A short form syntax exists to achieve such</p>

syntax: <b>newlist = [expression for item in iterable if condition == True]</b>

```python
my_list = [13,2, 3, 4,5, 6,71,8,91,101]
my_list2 = [] #odd values

for i in range(len(my_list)):
    elt = my_list[i]
    if elt % 2 == 1: #elt is odd
        my_list2.append(elt)
print("odd elts list:",my_list2)

my_list3 = [item for item in my_list if item%2== 1]
print("odd elts list:", my_list3)
```

```python
my_word = "This city is great"
my_words = my_word.split(" ") #this is a list
elts_count = [len(item) for item in my_words]
print(elts_count)
```

```python
my_list4 = [item+1 for item in range(10)]
print(my_list4)
```

```python
#.append() = add an element at the end of a list
#.extend() = add a list to a list

first_list = [1,3,4]
second_list = [5,3]

third_list = first_list+second_list

first_list.extend(second_list)
print(first_list)
print(third_list)
```

```python
list1 = [1,2,3] #we have created an object 'L1'
#We have not created an object, we have simply copied the location of list1
list3 = list1.copy() #we have created a new object

list2 = list1 
list2[0] = 8 #We have modified list1 using the reference variable "list2"

print(list1)
print(list2)
print(list3)
```

```python
my_list = [1,2,4,5,9,10]

print(my_list)
my_list.pop(3)#removing the 4th element (index = 3)

print(my_list)
```

```python
my_list = [1,2,4,5,9,10]

print(my_list)
del my_list[3]#removing the 4th element (index = 3)

print(my_list)

#my_list.clear()
my_list = []

print(my_list)
```

<h3>Lesson 2: A Tuple</h3>
<p>A tuple is collection of data types organised in orderly fashion</p>
<p>Unlike Lists, <b>Tuples are immutable objects</b>, that is its content cannot be modified </p>
<p>Functions and methods to remember: <b>len()</b></p>

<img src="media/tuple_interface.png" width="600px"/>

```python
my_tuple = (3,4,5)
my_list = list(my_tuple)
my_list[0] = 4
my_tuple = tuple(my_list)

print(my_tuple)
```

```python
def list_summary(my_list):
    f_value = my_list[0]
    l_value = my_list[-1]
    result = (f_value,l_value)
    return result

my_list = [3,4,5,8]

result = list_summary(my_list)
print(result)
```

<h3>Lesson 3: Dictionnary</h3>
<p>A dictionary is a collection of data types stores by means of unique identifiers called <b>"keys"</b></p>
<p>A Dictionary is similar to a list, however, data are accessed by <b>keys</b> rather than <b>indexes</b></p>
<p><b>car_description[0] = 'mazda'</b>  (using a list)</p>
<p><b>car['brand'] = 'mazda'</b>  (using a dictionary)</p>
<p>It is appropriate to use a dictionary when storing an object with different properties</p>
<p>e.g. <b>soccer_player = {"name":"lionel", "surname":"messi", "age":34}<b/></p>
<p><b>player_name = soccer_player['name']</b></p>

```python
my_list = ["kim", "msibi", 20, 1.8]
print("name:", my_list[0])
print("surname:", my_list[1])
print("age:",my_list[2])
print("height:",my_list[3])
```

```python
my_dict = {"name":"kim", "surname":"msibi", "age":20, "height":1.8}
print("name:", my_dict["name"])
print("surname:", my_dict["surname"])
print("age:",my_dict["age"])
print("height:",my_dict["height"])
```

```python
for key in my_dict:
    elt = my_dict[key]
    print(key,":", elt)
```

```python
for key,value in my_dict.items():  #for value in my_dict.values(): 
    print(key,":",value)
```

<h3>Exercise 1</h3>
Write a Python function to sum all the items in a list.

```python
def sum_list(my_list):
    N = len(my_list)
    sum_value = 0
    for i in range(N):
        sum_value = sum_value + my_list[i]
    return sum_value


my_list = [20,34,56,78]
sum = sum_list(my_list)

print("The sum is",sum)
```
