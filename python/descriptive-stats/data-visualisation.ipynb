```python
import warnings 
warnings. filterwarnings('ignore')

import pandas as pd

df = pd.read_csv('datasets/Healthcare-Diabetes.csv') #loading data into a dataframe
def age_grouping(age):
    if age < 20:
        group = 'teenager'
    elif age < 40:
        group = 'adult'
    else:
        group = 'old-person'
    return group

df['age-group'] = df['Age'].apply(age_grouping)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.boxplot(df['Age'])
plt.title('Box plot - Age variable')
plt.show()
```

```python
count_values = df['age-group'].value_counts().tolist()
```

```python
import matplotlib.pyplot as plt


#df['age-group'].value_counts()

plt.figure()
plt.pie(count_values,labels=["adult","old-person"])
plt.title('Data distribution in qualitive variable: Age-group')
plt.show()
```

```python
import numpy as np

x = df['Glucose']
y = df['BloodPressure']

a,b = np.polyfit(x,y,1)

plt.figure()
plt.scatter(df['Glucose'],df['BloodPressure'])
plt.plot([x.min(),x.max()],[a*x.min()+b,a*x.max()+b],color="r")
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.title('BP = %.2fG + %.2f'%(a,b))
plt.show()
```
