```python
import matplotlib.pyplot as plt
import numpy as np

grade_vs_q1 = sub_df.iloc[:,[0,10]]
grade_group = grade_vs_q1.grade.unique().tolist()
grade_group

grade_vs_q1_df = grade_vs_q1[grade_vs_q1['grade']=='AVP'].value_counts()
grade_bar_data = grade_vs_q1_df.AVP.tolist()
grade_bar_data = np.array(grade_bar_data)
grade_bar_data = grade_bar_data/grade_bar_data.sum()

plt.figure()
plt.bar(grade_vs_q1_df.AVP.index.tolist(),grade_bar_data)
plt.title(grade_vs_q1.columns[1])
plt.xlabel('AVP')
plt.ylabel('Frequency')
plt.show()
```
