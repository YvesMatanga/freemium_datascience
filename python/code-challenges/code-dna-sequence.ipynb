```python

def code_seq_comp(seq_in):
    
    seq_out =''
    N = len(seq_in)#get the len of the functions   
    for i in range(N):
        seq_val = seq_in[i]#fetch a sequence value
        comp_val = ''
        if seq_val == 'A':
            comp_val = 'T'
        elif seq_val == 'T':
            comp_val = 'A'
        elif seq_val == 'G':
            comp_val = 'C'
        else:
            comp_val = 'G'
            
        seq_out+=comp_val
    return seq_out
```

```python

seq = 'TAGCACCATG'
com_seq = code_seq_comp(seq)
print("seq:",seq, "comp-seq:",com_seq)
```
