```python
def symmetric_sum(list_):
    n = len(list_)
    
    start_index = 0
    end_index = n-1
    
    while start_index <= end_index: 
        left_val = list_[start_index]
        right_val = list_[end_index]
        
        if left_val == right_val:
            break
            
        start_index = start_index+1
        end_index = end_index-1
        
    
    sub_list = list_[start_index:end_index+1]
    
    return sub_list,sum(sub_list)
    
```

```python
list1 = [9,8,7,5,7,8,12]
sub_list,sum_sb = symmetric_sum(list1)

print(sub_list,sum_sb)
```

```python
def pascal_triangle(c_depth):
    depth = c_depth
    llist = []    
    for li in range(depth):
        if li < 1:
            llist.append([1])
        else:
            cur_list = llist[li-1] #get previous list
            #print("in - depth:", li,cur_list)
            new_list = []
            n_cur_list = len(cur_list)
            #add in between values
            for i in range(len(cur_list)):
                if i < n_cur_list-1:
                    new_list.append(cur_list[i]+cur_list[i+1])
                    
            #add boundary values
            new_list.insert(0,1)
            new_list.append(1)
            #add new list to the tree
            #print("out - ",new_list)
            llist.append(new_list)
    
    #print the values
    for i in range(depth):
        list_i = llist[i]
        for k in range(depth-i):
                print("  ",end="")
        for j in range(len(list_i)):                
            print(list_i[j]," ",end="")
        print("")
```

```python
pascal_triangle(8)
```
