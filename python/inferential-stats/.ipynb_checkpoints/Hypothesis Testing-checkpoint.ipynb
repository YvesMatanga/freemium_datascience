<h3>Hypothesis Testing</h3>

<p>Hypothesis Testing is a test to verify the significance of a claim made on the population</p>
 Assuming a given claim is made on the population. The attempt of hypothesis testing is to verify whether on a typical sample size on the population, the claim will be consistent.In other words, if there is significant evidence to reject the hypothesis.

<h4>1.1 The significance test</h4><br/>
The significance test is a way of statistically testing a hypothesis by comparing the data to values predicted by the hypothesis. <b>Data that fall far from
the predicted values provide evidence against the hypothesis.</b>

Elements of the significanc test:

<ul>
    <li>The <b>null hypothesis: $H_0$</b>: It is the claim that is being made about the population</li>
    <li>The <b>alternative hypothesis $H_a$</b>: It is the negatiion of the null hypothesis</li>
    <li>The <b>test statistic</b>: It is the numerical claim about the data</li>
    <li>The <b>p-value</b>: The probability threshold of the amount of evidence against the null</li>
    <li>The <b>sampling distribution</b>: It is the known (or assumed) sampling distribution of the test statistic</li>
    </ul>

<h4>1.2 The significance test for a population mean</h5>

<b>Assumption</b>

Let a population variable X be normally distributted with the mean $\mu$ unknown and standard deviation $\sigma$ known.


<b>Hypotheses</b>

$$H_0: \mu = \mu_0$$

$$H_a : \mu \neq \mu_0$$

<b>Test statistic (i.e. verification phase)</b>

Compute the sample mean $\bar{X}$ from the data. If $H_0: \mu=\mu_0$ is true, then the center of the sampling distribution should be $\mu_0$. 

The evidence about $H_0$ is the distance of the sample value $\bar{X}$ from the null hypothesis value $\mu_0$, relative to the standard error. An observed value
of $\bar{X}$ falling far out in the tail of this sampling distribution of $\bar{X}$ casts doubt
on the validity of H0, because it would be unlikely to observed value $\bar{x}$ of $\bar{X}$
very far from μ0 if truly μ = μ0.

The test statistic is normalised

$$ Z = \frac{\bar{X}-\mu_0}{\sigma/\sqrt{n}}$$

<b>p-value</b>

The p-value is the probability of obtaining a z value away from the assumed population mean if $H_0$ was true.


$$ 2.P(Z\geq |z|)=p $$ Z is $N(0,1)$

<h4>Signicance Test process</h4><br/>
Null Hypothesis: $H_0: \mu = \mu_0$<br/>
Alternate Hypothesis: $H_0: \mu \neq 30$<br/>
Test statistic: $\bar{X}$<br/>
Sampling distribution: Identify the nature of the sampling distribution

Compute $z=\frac{\bar{X}-\mu_0}{(\frac{\sigma}{\sqrt{n}})}, Z\approx N(0,1)$

Find $p = 2P(Z>|z|) = 2P(Z < -|z|) = 2CDF_Z(-|z|)$<br/>
Define the <b>significance level</b> $\alpha$ (i.e. a typically small number, $\alpha=0.05$).<br/>
if p > $\alpha$, accept the null hypothesis<br/>
if p $\leq \alpha$, reject the null hypothesis, in favour of the alternate hypothesis



<b>Exercise - Hypothesis</b>


<p>The population mean of the random process below is claimed to be 30. Verify the reliability of this 
claim at </p> 

```python
data = [30,32,21,43,32,36,37,31,30,32, 21, 3,32,21,43,32,36,43,32,36,32,21,43,32,36,43,32,36,32,21,43,32,36,43,32,36]
```


Null Hypothesis: $H_0: \mu = 30$<br/>
Alternate Hypothesis: $H_0: \mu \neq 30$<br/>
Test statistic: $\bar{X}$
Sampling distribution: $ N(\mu, \frac{\sigma}{\sqrt{n}})$

Compute $z=\frac{\bar{X}-\mu_0}{(\frac{\sigma}{\sqrt{n}})}, Z\approx N(0,1)$

Find $p = 2P(Z>|z|) = 2P(Z < -|z|) = 2CDF_Z(-|z|)$

if p > $\alpha$, accept the null hypothesis<br/>
if p $\leq \alpha$, reject the null hypothesis, in favour of the alternate hypothesis



```python
## import numpy as np
import scipy.stats as st
import numpy as np

n = len(data)
pop_mean = 30
std_pop_mean = np.array(data).std() #n is sufficiently large, and thus it can be approximated from the sample
alpha = 0.05

sampling_distrib_std = std_pop_mean/np.sqrt(n)
sample_mean = np.array(data).mean()

test_stat_norm = (sample_mean - pop_mean)/sampling_distrib_std

cdf_z_neg = st.norm.cdf(-np.abs(test_stat_norm),loc=0,scale=1)
p_val = 2*cdf_z_neg

print('p-val: 2*P(Z>|z|):',p_val)

if p_val <= alpha:
    print('The null hypothesis is rejected')
else:
    print('The null hypothesis is accepted')    
```

<b>Other Statistical tests</b>

<p>Several significance tests exists depending on the hypothesis that needs testing<p>
   <ul>
       <li><b>Equality of two means<b/>:<br/> $H_0: \mu_0=\mu_1$</li>
       <li><b>ANOVA</b> test - Analysis of variance: compares whether independent variables share the same mean: $H_0: \mu_0=\mu_1=\mu_2=..=\mu_n$  </li>
       <li><b>Testing equality of variances:</b> <br/>$H_0: \sigma_0 = \sigma_1=...=\sigma_n$</li>
       <li>etc.</li>
   </ul>
