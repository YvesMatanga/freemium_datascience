```python
my_list = [item+1 for item in range(10)]
print(my_list)
```

```python
def sum_list(my_list_):
    sum_ = 0
    n = len(my_list_)
    for i in range(n):
        sum_ = sum_ + my_list_[i]  #To add on top of what you had 
    return sum_
```

```python
print("The sum is :",sum_list(my_list))
```

```python
def find_min(list_):  #find the minimum value in the list
    min_value = float('infinity')
    n = len(list_)
    
    for i in range(n):
        if list_[i] < min_value:
            min_value = list_[i]  # = : assign   # ==: is it equal? (boolean )
    
    return min_value    
```

```python
my_list2 = [10, 8, 9, -1, 9]
min_val = find_min(my_list2)
print("The minimum value is ",min_val)
```

```python
def find_max(list_):  #find the minimum value in the list
    max_value = -float('infinity')
    n = len(list_)
    
    for i in range(n):
        if list_[i] > max_value:
            max_value = list_[i]  # = : assign   # ==: is it equal? (boolean )
    
    return max_value  
```

```python
max_val = find_max(my_list2)
print("The maximum value is ",max_val)
```

```python
def bubble_sort(list_p):
    list_t = list_p.copy()
    n = len(list_t)
    
    for rounds in range(n):#swapping rounds
        for i in range(n-1):
            if list_t[i] > list_t[i+1]:
                list_t[i],list_t[i+1]  = list_t[i+1],list_t[i] #swap
    
    return list_t      
        
```

```python
list3 = [30, 4, 8, 7,-1]
list_sorted = bubble_sort(list3)
print(list_sorted)
```

```python
list_sorted2 = sorted(list3,reverse=True)

print(list_sorted2)
```
