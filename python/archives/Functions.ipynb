```python
s_count = 0
number = int(input('Enter a pos number:'))
s_count += 1 #Add + 1
while number < 0:
    number = int(input('Enter a pos number:'))
    s_count += 1 #Add + 1
    
print("You have done it (", number, ") after ", s_count, "times")
```

<h4>The Function operation requires two main activities<h4>
    <ol><li>Function definition</li>
        <li>Function call</li>
    </ol>

f: R -----> R  #function defition
   x------> f(x) = 2x + 1
   
   
  f(3) = 2x3+1 = 7 #function call

Four types of functions base on the existence or absence of the input/ouput

(1) = Input / Ouput  e.g c = add(1,2) 
(2) = Input / No ouput e.g. init(options)
(3) = No Input / Output e.g  rand()
(4)  = No Input/ No Ouput e.g menu()

```python


"""
def function_name(input arguments):
    #Function definitio
    
    #Return the value(s)
"""  

def add(num1,num2):  #Function definition
    result = num1+num2
    return result

def student_profile(std_num, status): #Input/ NO (return) output
    print("*"*30)
    print("| Student number:", std_num)
    if status == 1:
        status_mess = "Enrolled"
    else:
        status_mess = "Pending"
    print("| Status: ", status_mess)
    print("*"*30)
    
import random

def generate_random_age():
    age = random.randint(0, 120)
    return age

def menu():
    print("*"*30)
    print("| Welcome to Burger King")
    print("*"*30)
    print("\t| 1. Burger Only")
    print("\t| 2. Meal")
    print("\t| 3. Free Drinks")
    print("*"*30)
    print("\n\tMake your Choice: ") #\n = newline \t = tab
```
