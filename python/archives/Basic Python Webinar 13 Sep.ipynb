<h1>Basic Arithmetic</h1>

```python
number1 = 20
number2 = 3

sum = number1 + number2 #add
diff = number1 - number2#subtract
quotient = number1 / number2#divide
product = number1*number2  #multiply

rem = number1%number2 #finding the remainder of integer division
int_div = number1//number2 #integer division
power = number1**number2 #number1 to the power of number 2

print(f"{number1} + {number2} = {sum}")
print(f"{number1} - {number2} = {diff}")
print(f"{number1} * {number2} = {product}")
print(f"{number1} / {number2} = {quotient}")
print(f"{number1} mod {number2} = {rem}")
print(f"{number1} // {number2} = {int_div}")
print(f"{number1}^{number2} = {power}")
```

<p>Create a program that calculates someone's body mass index and returns the person's status</p>
$$
BMI = \frac{mass (Kg)}{height^2 (m^2)}
$$

            <td> Between 18.5 and 24.99 </td><td> Normal</td>

        <tr>

            <td> $\geq 29.99$ </td><td> Obese</td>
        </tr>

<p><b>Demo</b></p>
<p>Enter your mass (Kg): 80</p>
<p>Enter your height (m): 1.8</p>
<p>Your body mass index is <b>24.69</b> and you are <b>normal</b></p>


```python
mass = float(input('Enter your mass (Kg)'))
height = float(input('Enter your height (m)'))
bmi = mass/(height*height)


if bmi < 18.5:
    category = "underweight"
elif bmi < 24.99:
    category = "normal"
elif bmi < 29.99:
    category = "overweight"
else:
    category = "obese"

print(f"Your BMI is {round(bmi,2)} - Status: {category}")
```

<h5>Sentinel Controlled loop: while loop -- It stops based on a termination criterion -- termination iteration: unknown</h5>
<h5>Counter Controlled loop:  for loop -- It stops based on a termination count -- termination iteration: known</h5>

```python
#Sentinel controlled loop
# while (boolean statement is true):
#    keep on executing this code

store_count = 0
c_number = int(input('Enter a positive number:'))
while c_number < 0:
    c_number = int(input('Enter a positive number:'))
    store_count = store_count + 1

print(f"Finally after {store_count} iteration")


    
```

<h3>Exercise 7</h3>
<p>Create a function that finds the factorial of a given number</p>
<p>factorial(5) = 5 x 4 x 3 x 2 x 1 = 120</p>
<p>factorial(4) = 4 x 3 x 2 x 1 = 24 </p>

```python
def factorial(n):
    init_val = n
    product = init_val
    for i in range(n-1):#(n-1) repetitions
        init_val = init_val - 1
        product = product*init_val
    return product

    
```

```python
def factorial2(n):
    product = 1
    for i in range(1,n+1):#(n-1) repetitions
        product = (i)*product
    return product
```
