```python
#Sentinel controlled loop
# while (boolean statement is true):
#    keep on executing this code

store_count = 0
c_number = int(input('Enter a positive number:'))
store_count = store_count + 1

while c_number < 0:
    c_number = int(input('Enter a positive number:'))
    store_count = store_count + 1

print(f"Finally after {store_count} iteration(s)")   
```

<h3>Counter Controlled loop</h3>

```python
#for loop
# for index in range(total_count):
#   run this code

for i in range(10):
    print(f"{i} - Hello Everyone")
    
print("-- Done---")
```

```python
i = 0 
while i < 10:
    print(f"{i} - Hello Everyone")
    i = i + 1
print("---Done---")
```
