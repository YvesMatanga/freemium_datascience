<h3> Question 2 - Simple Linear Regression - Optimising crops production: Merka Agri</h3><br/><br/>

Merka Agri is a wholesaler producer of corns. The company makes use of fertilizer to accelerate the growth of its crops added for 90 days period from planting prior to the harvesting. The amount of fertilizer obviously as an impact of the growth of the corns and the company would like to have an adequate estimate of the relationship between the amount of fertilizer added overall against the weight of the final corn. You have conducted an experiment for 100 crops whereby a fixed amount of mg per pot is added every day and at the end of 90 days the weights of the corns has been recorded. You are currently in the analysis of the data (“merka_argi_corn_experiment.csv”).



<h3 style="color:blue">Solution steps </h3>

<h4 style="color:blue">Step 1: load the datasets using pandas. List all attributes, identify the feature (x) and the target (y)</h4>

tip: df.info(), df.describe()

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('db/merka_agri_corn_experiment.csv')
df.head()

X = df.iloc[:,]
```

```python
X = df.iloc[:,[1]] #feature(S)
y = df.iloc[:,[-1]] #target
```

```python
df.info()
```

<h4 style="color:blue">Step 2: Generate a scatter plot - x-axis: total-mg-per-pot, y-axis:corn-weight</h4>

```python
import matplotlib.pyplot as plt
import numpy as np 

plt.figure()
plt.scatter(X,y, color="blue")
plt.xlabel('fertiliser-weight')
plt.ylabel('corn-final-weight')
plt.title('Merka Agri Corn production experiment')
plt.show()
```

<h4 style="color:blue">Step 3: Build a linear regression model and extract the model coefficients</h4>

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X,y)

y_pred = lm.predict(X)
```

<h4 style="color:blue">Step 4: Compute the R^2 score and RMSE based on the model estimation as well provide a goodness of fit scatter plot</h4>

```python
y_ = y.to_numpy().reshape(1,-1)[0]
y_pred_ = y_pred.reshape(1,-1)[0]
```

```python
from sklearn import metrics
import math 
import matplotlib.pyplot as plt
import numpy as np

r2_score = metrics.r2_score(y,y_pred)
rmse = math.sqrt(metrics.mean_squared_error(y,y_pred))

a,b = np.polyfit(y_,y_pred_,1) #y = ax+ b
min_x = y.min()
max_x = y.max()

plt.figure()
plt.scatter(y,y_pred,color="blue")
plt.plot([min_x,max_x],[min_x*a+b,max_x*a+b],color="red")
plt.title('R2 score: %.3f - RMSE: %.3f'%(r2_score,rmse))
plt.show()
```

<h4 style="color:blue">Step 6: Comments on whether your model performs well or not?</h4>

<h4 style="color:blue">Step 7: On average, what could the weight of a corn if fertalised with 50-mg for 90days?</h4>

```python
x_un = 50
y_un = lm.predict(np.array([[x_un]]))

print('Weight - prediction for 50 mg:',round(y_un[0][0],3), '')
```

<h4 style="color:blue">Step 2: Identify records containing missing valueds and remove them</h4>

<h4 style="color:blue">Step 3: Extract Car Band from the dataset</h4>

Tips: 
<ul>
    <li>df['CarName'] = df['CarName'].apply(define_string_manip_func1)</li>
</ul>

```python
df[['CarName']]
```

```python
def transform_rule(str_in):
    return str_in.split(' ')[0];

df['CarName'] = df['CarName'].apply(transform_rule)
df.head()
```

<h4 style="color:blue">Step 4: Identify features (X) and the target variable (y)</h4>

```python
df.columns
```

```python
Nf = len(df.columns)
X = df.iloc[:,2:-1]
X.head()
```

```python
y = df.iloc[:,[Nf-1]]
y.head()
```

<h4 style="color:blue">Step 5: Convert categorical features into numerical equivalent attributes</h4>

```python
X_mod = pd.get_dummies(X,drop_first=True)
X_mod.head()
```

<h4 style="color:blue">Step 6: Perform Correlation analysis using a pairplot and a correlation matrix on the feature data matrix</h4>

Use your own judgment to remove some features (i.e. Do not remove any for now. You may do so after evaluating your model performance on an all-feature training first round)

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = X_mod.corr()#correlation matrix
sns.heatmap(cm, annot=True)
plt.title('Correlation analysis')
plt.show()
```

```python
cm
```

<h4 style="color:blue">Step 6b:Scale your features</h4>

```python
from sklearn.preprocessing import StandardScaler 

Sc = StandardScaler()
X_mod_sc = Sc.fit_transform(X_mod)
X_mod_sc_df = pd.DataFrame(X_mod_sc,columns = X_mod.columns)
X_mod_sc_df.head()
```

<h4 style="color:blue">Step 7: Split the dataset in training/test set using a 80/20 split proportion</h4>

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test  = train_test_split(X_mod_sc_df, y, test_size = 0.2, random_state=1234)
```

<h4 style="color:blue">Step 8: Build a linear regression model and extract the model coefficients</h4>

```python
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
mlr.fit(X_train, y_train)#training the model

y_pred_train = mlr.predict(X_train)
y_pred_test = mlr.predict(X_test)
```

<h4 style="color:blue">Step 9: Compute the $R^2$ score and RMSE based on the model estimation as well provide a goodness of fit scatter plot - on the training set and test set separately</h4>

```python
from sklearn import metrics
import math

r2_score_train = metrics.r2_score(y_train,y_pred_train)
r2_score_test = metrics.r2_score(y_test,y_pred_test)

rmse_train = math.sqrt(metrics.mean_squared_error(y_train,y_pred_train))
rmse_test = math.sqrt(metrics.mean_squared_error(y_test,y_pred_test))


plt.figure()
plt.xlabel('observation')
plt.ylabel('prediction')
plt.scatter(y_train,y_pred_train)
plt.title('Training set R2:%.3f RMSE: %.3f'%(r2_score_train,rmse_train) )
plt.show()

plt.figure()
plt.xlabel('observation')
plt.ylabel('prediction')
plt.scatter(y_train,y_pred_train)
plt.title('Test set R2:%.3f RMSE: %.3f'%(r2_score_test,rmse_test))
plt.show()
```
