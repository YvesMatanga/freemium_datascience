<h2>Case-Study 2: Nonlinear Regression using Decision Trees</h2>

Design a Salary Prediction model for a Baseball player playing in major leagues, given player profile and game performance attributes. This dataset was originally taken from the StatLib library which is maintained at Carnegie Mellon University. This is part of the data that was used in the 1988 ASA Graphics Section Poster Session. The salary data were originally from Sports Illustrated, April 20, 1987. The 1986 and career statistics were obtained from The 1987 Baseball Encyclopedia Update published by Collier Books, Macmillan Publishing Company, New York.

```python
df_bb = pd.read_csv('db/Hitters.csv')
df_bb.head()
```

<b style="color:blue;">Step 1: Drop records with missing values</b>

```python
df_bb.dropna(inplace=True)#remove nan
```

<b style="color:blue;">Step 2: Using Mutual Information, draw a barplot of each feature importance in descending order of importance</b>.( Use HotEncoder, OrdinalEncoder dependending on the case)

```python
from sklearn.feature_selection import mutual_info_regression
df_bb_enc = pd.get_dummies(df_bb)#one hot encoding on categorical variable

X = df_bb_enc.drop(columns=['Salary'])
y = df_bb_enc[['Salary']]

#compute mutual information 
MI_scores = mutual_info_regression(X,y)

MI_score_df = pd.DataFrame({'features':X.columns, 'MI':MI_scores})

MI_df_sorted = MI_score_df.sort_values(by='MI',ascending=False)
MI_df_sorted.head()

plt.figure()
plt.barh(MI_df_sorted['features'],MI_df_sorted['MI'])
plt.gca().invert_yaxis()
plt.show()

```

```python
feat_perc = 0.6
N = int(len(X.columns)*feat_perc)


fSubset = MI_df_sorted['features'][0:N].tolist()
fSubset

X_filtered = X.loc[:,fSubset]
X_filtered.head()
```

<b style="color:blue;">Step 3: Select the N more important features.</b>

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.8, random_state=1234)
```

<b style="color:blue;">Step 4: Split the dataset in 80/20 training/test set proportion</b>

<b style="color:blue;">Step 5: Build the decision tree with proper parameter configuration and train the model with full_features and with filtered features</b>

```python
from sklearn.tree import DecisionTreeRegressor

dt_1 = DecisionTreeRegressor(max_depth=None, min_samples_leaf=1)
dt_2 = DecisionTreeRegressor(max_depth=None, min_samples_leaf=1)

X_train_f = X_train.loc[:,fSubset]
X_test_f = X_test.loc[:,fSubset]

dt_1.fit(X_train,y_train)#every featyre
dt_2.fit(X_train_f,y_train)#some feature
```
