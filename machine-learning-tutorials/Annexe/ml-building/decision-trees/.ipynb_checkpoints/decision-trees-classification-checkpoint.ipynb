```python
class cart_classifier():
    #CART classifier
    
    def __init__(self):
        self.X_d = [] 
        self.y_d = []
    
    def fit(self,X,y):
        self.X_d = X
        self.y_d = y
        
        return -1
    
    def gini_index(self,X,y,f_i): #get the gini score for the data
        Xi_d = X.iloc[:,f_i]  #get feature
        
        if Xi_d.dtype != 'float': #categorical data type
            uniq_vals = Xi_d.unique() #get unique values in the categorical feature vector
            return -1
        else: #numeric data tye
            return -1
        
```

```python
import pandas as pd
df = pd.read_csv('../datasets/classification/golf_session_data.csv')
df.head()
X = df.iloc[:,0:len(df.columns)]
y = df.iloc[:,[len(df.columns)-1]]
```
